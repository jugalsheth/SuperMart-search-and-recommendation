```python
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('superstore_final.csv',encoding= 'unicode_escape')
```

```python
df.columns
```

```python
df.head()
```

```python
df['Product_Price']=df['Sales']/df['Quantity']
```

```python
df_price=df[['ProductID','Product_Price']].groupby(['ProductID']).agg('mean')
```

```python
df_price
```

```python
df_category=df.groupby(["ProductID",'Category']).count().reset_index()[['ProductID','Category']]
```

```python
df_category
```

```python
df_frequent=df[['ProductID','Customer ID']].groupby(['ProductID']).agg('count')
```

```python
df_frequent.columns=['Count']
```

```python
df_frequent
```

```python
df_mostrated=df[['ProductID','Rating']].groupby(['ProductID']).agg('mean')
```

```python
df_mostrated
```

```python
df_popular=pd.merge(df_frequent,df_mostrated,left_index=True, right_index=True)
```

```python
df_popular=pd.merge(df_popular,df_price,left_index=True, right_index=True)
```

```python
df_popular
```

```python
df_popular=pd.merge(df_popular,df_category,left_index=True, right_index=True)
```

```python
df_popular=df_popular.sort_values(by=['Count','Rating'], ascending=False)
```

```python
df_popular.head(10)
```

```python
df_popular.to_csv('df_popular.csv')
```
